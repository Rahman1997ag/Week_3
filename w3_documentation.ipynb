## SCENARIO

A student has marks for three quizzes.

### Create a Python program that:
1. Ask the user to enter three quiz marks.
2. Calculate the average mark.
3. Display the average.
4. Determine whether the student passes or fails.
5. Allow another student's marks to be entered.

---

## 1. Identify the Components

### 1.1. Inputs
- Three quiz marks (floating point numbers).
- User's choice to continue (yes/no or y/n).

### 1.2. Process
- Read three marks.
- Compute the average: (mark1 + mark2 + mark3) / 3.
- Compare average against passing threshold (>= 50).
- Repeat based on user input.

### 1.3. Outputs
- The average mark.
- Pass or Fail status.
- Program termination message.

## 2. Design the Algorithm

### 2.1. Pseudocode

```
START
    choice = "y"
    WHILE choice == "y" DO
        INPUT quiz1, quiz2, quiz3
        average = (quiz1 + quiz2 + quiz3) / 3
        OUTPUT average
        IF average >= 50 THEN
            OUTPUT "Pass"
        ELSE
            OUTPUT "Fail"
        ENDIF
        INPUT choice (continue? y/n)
    ENDWHILE
    OUTPUT "Program Ended"
END
```

### 2.2. Identify the Algorithm Structure

#### 2.2.1. Sequence
- The program executes steps in order: input, calculation, output, decision, and loop control.
- Within each iteration, the sequence is: get marks, compute average, display average, check pass/fail, ask for continuation.

#### 2.2.2. Selection
- The IF statement selects between two outcomes based on the condition `average >= 50`.
- If true, the student passes; otherwise, fails.

#### 2.2.3. Iteration
- The WHILE loop repeats the entire block of statements as long as the user enters 'y' (or 'yes') to continue.
- This allows multiple students' marks to be processed.

## 3. Python Program (Student_Ave_Calc.py)

The following code implements the algorithm described above.

````python
def Evaluate(average):
    if average >= 50:
        return "SUCCESS"
    else:
        return "FAILED"

choice = "yes"
while choice == "yes":
    firstMark = float(input("Please enter first quiz mark(%100)  : "))
    secondMark = float(input("Please enter second quiz mark(%100) : "))
    thirdMark = float(input("Please enter third quiz mark(%100)  : "))

    average = float((firstMark + secondMark + thirdMark) / 3)

    print(f"THE AVERAGE IS : {average}")
    print(f"YOU ARE        : {Evaluate(average)}")
    choice = input("Do you want to try again ? (yes/no) : ")

print("Program Ended")
````
